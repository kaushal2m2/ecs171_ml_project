```python
import pandas as pd
import numpy as np
import os
import sys

df = pd.read_csv(os.path.join(sys.path[0], "healthcare-dataset-stroke-data.csv"))
df = df.drop(columns=['id'])
```

```python
def replace_smoke(type):
    if (type == 0): #remove all unknow row
        df.replace(["never smoked", "formerly smoked", "smokes"],[0,1,2],inplace=True)
        df.drop(df[df['smoking_status'] == 'Unknown'].index, inplace=True)
    if (type == 1): #set unknow as single category
        df.replace(["never smoked", "formerly smoked", "smokes", "Unknown"],[0,1,2,-1],inplace=True)
    if (type == 2): #default unknow to be smokes 
        df.replace(["never smoked", "formerly smoked", "smokes", "Unknown"],[0,1,2,2],inplace=True)

replace_smoke(0)
```

```python
df = pd.get_dummies(df, columns = ["gender", "ever_married", "work_type", "Residence_type"], dtype=int) 
df = df.drop(columns="gender_Other")
print(df)
```

```python
def fill_NA_bmi(type):
    if(type == 0):
        bmi_mean = df["bmi"].mean()
        df['bmi'] = df['bmi'].fillna(bmi_mean)
    if(type == 1):
        bmi_median = df["bmi"].median()
        df['bmi'] = df['bmi'].fillna(bmi_median)
    if(type == 2):
        bmi_mode = df["bmi"].mode()
        df['bmi'] = df['bmi'].fillna(bmi_mode)
fill_NA_bmi(0)
```
